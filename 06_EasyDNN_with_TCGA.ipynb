```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adam


%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
X = pd.read_csv("/content/tcga_1000.csv",index_col=0)
Y = pd.read_csv("labels.csv",index_col=0)
```

```python
from sklearn.preprocessing import LabelEncoder

y_ordinal = LabelEncoder().fit_transform(Y)
Y_onehot = pd.get_dummies(y_ordinal)
Y_onehot
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y_onehot, test_size=0.3, random_state=138)
```

```python
#순차적인 모델을 만들것이다. 위에서부터 아래로 쌓는다.
model = Sequential()

#Dense Layer : 모든 뉴런마다 인풋이 전부 연결된 레이어
#Output 1개, input도 1개
model.add(Dense(32, input_shape=(1000,), activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(5, activation ='softmax'))

model.summary()
```

```python
model.compile(optimizer = Adam(lr=0.0001), loss = 'categorical_crossentropy', metrics=['acc'])

model.fit(X_train, Y_train, epochs=50, batch_size=64, validation_split=0.2)
```

```python
Y_pred = model.predict(X_test)
Y_pred = Y_pred.round()
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

print(accuracy_score(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
```
