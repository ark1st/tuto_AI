```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD


%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
y = ["장려", "우수", "우수", "최우수", "장려"]

df = pd.DataFrame({
    "x1" : x1_data,
    "x2" : x2_data,
    "x3" : x3_data,
    "y" : y
})

df
```

```python
from sklearn.preprocessing import LabelEncoder

x_train = df[['x1', 'x2', 'x3']]
y_train = df[['y']]


y_train_ordinal = LabelEncoder().fit_transform(y_train)
y_train_onehot = pd.get_dummies(y_train_ordinal)
print(y_train_onehot)
```

```python
#순차적인 모델을 만들것이다. 위에서부터 아래로 쌓는다.
model = Sequential()

#Dense Layer : 모든 뉴런마다 인풋이 전부 연결된 레이어
#Output 1개, input도 1개


model.add(Dense(3, input_shape=(3,), activation = 'softmax'))

model.summary()
```

```python
model.compile(optimizer = SGD(lr=0.01), loss = 'categorical_crossentropy', metrics=['acc'])

model.fit(x_train, y_train_onehot, epochs=300, batch_size=1)
```
